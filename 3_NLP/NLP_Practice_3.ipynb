```python
import nltk
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk import pos_tag
import re
text = "The striped bats were hanging on their feet and ate best fruits"

text_LowerCase = text.lower()
text_RemovePunctuation = re.sub(r'[^\w\s]', "", text_LowerCase) # word which is not in \w\s will replace with "" empty string
text_Tokenizing = word_tokenize(text_RemovePunctuation) # Tokenizing
#text_RemoveStopwords = [word for word in text_Tokenizing if word not in stopwords.words('english')]
text_FeatureExt_FreqCount = nltk.FreqDist(text_Tokenizing) # Counting repeated words
text_RemoveDublicates = list(text_FeatureExt_FreqCount.keys()) # Converting the dict to list
text_PosTagging = pos_tag(text_RemoveDublicates) # POS tagging
text_PosTagging
```

```python
def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return 'a'
    if tag.startswith('V'):
        return 'v'
    if tag.startswith('N'):
        return 'n'
    if tag.startswith('R'):
        return 'r'
    else:
        return 'n' # default to noun
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemmatizer.lemmatize('running', pos='v')
```

```python
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
lemmatizer = WordNetLemmatizer()
def pos_tag_and_lemmatize(text):
    words = word_tokenize(text)
    pos_tags = pos_tag(words)
    
    lemmas = []
    for word, tag in pos_tags:
        lemma = lemmatizer.lemmatize(word, pos=get_wordnet_pos(tag))
        lemmas.append((word, tag, lemma))
    return lemmas
text = "The striped bats were hanging on their feet and ate best fruits"
result = pos_tag_and_lemmatize(text)
result
```
