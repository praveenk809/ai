```python
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
```

```python
text = "Very orderly and methodical he looked, with a hand on each knee, and a loud watch ticking a sonorous sermon under his flapped newly bought waist-coat, as though it pitted its gravity and longevity against the levity and evanescence of the brisk fire."
print(text)
```

```python
words = word_tokenize(text)
```

```python
wordnet_lemmatizer = WordNetLemmatizer()
wordnet_lemmatizer.lemmatize("driving", pos="v")
```

```python
wordnet_lemmatizer.lemmatize("happier", pos="a")
```

```python
lemmatinzed_words = []
for word in words:
    lemmatized_word = wordnet_lemmatizer.lemmatize(word, pos="v")
    lemmatinzed_words.append(lemmatized_word)

print(lemmatinzed_words)
```
