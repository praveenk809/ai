WordNet is a lexical database of English. It organizes words into synsets (sets of synonyms) and connects them via semantic relations.

```python
import nltk
from nltk.corpus import wordnet
```

```python
nltk.download('wordnet')
```

```python
synsets = wordnet.synsets("dog")
synsets
```

```python
Synonyms (synset): dog, domestic dog, Canis familiaris
Hypernym (is-a, more general): canine, canid → (dog is a kind of canine)
Hyponyms (is-a, more specific): working dog, toy dog, hunting dog
Meronyms (part-of): paw, tail, fur
Gloss (definition): "a domesticated carnivorous mammal that typically has a long snout, an acute sense of smell, and a barking voice"
```

```python
import requests

concept = "dog"

url = f"https://api.conceptnet.io/c/en/{concept}"

response = requests.get(url).json()
response
```

```python
for edge in response['edges']:
    rel = edge['rel']['label'] # relation type (IsA, capable of)
    start = edge['start']['label']
    end = edge['end']['label']
    print(f"({start}) - [{rel}] - ({end})")
```
