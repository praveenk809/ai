```python
# Write a regex to match strings that start with "Hello".

import re
text = "Hello, This is Praveen"
match = re.search("hello", text, re.IGNORECASE)
match.group()
```

```python
#Write a regex to match strings that end with "!"
text = [" This is Praveen!", " This is Naveen!", "this is Sunil"]

for i in text:
    match = re.search("!", i)
    if match:
        print(i)
```

```python
# Write a regex to match exactly "Yes" (and nothing else).

text  = ["Praveen lives in USA = Yes", "Dallas is in Florida = No", "Delhi is in India = Yes"]
for i in text:
    match = re.search('Yes+', i)
    if match:
        print(i)
```

```python
test = " yes, he is my frind. Yes, we go for tea"
re.findall("Yes*", test, re.IGNORECASE)
```

#Username validation: Must start with a letter and contain only letters, numbers, and underscores.
- "John_Doe" => Correct
- "123John" => Wrong

```python
users = ["praveen_g", "123Praveen", "suneel_j", "naveen_d40", "praveen1", "suneel1"]
for user in users:
    match = re.search("[a-zA-Z]+_+\w", user)
    if match:
        print(user)
```

```python
# Phone number check: Must be exactly 10 digits.

contacts = ["203-100-0000","0-1-2010", "408-32-876", "01947", "123-335-1099"]
pattern = ("^\d{3}-\d{3}-[0-9]{4}$")
for c in contacts:
    match = re.search(pattern, c)
    if match:
        print(c)
```

```python
contacts = ("203-100-0000","000-1-2010", "408-32-876", "01947", "123-335-1099" )

pattern = "^\d{3}-[0-9]{3,3}-\d+$"
for i in contacts:
    match = re.search(pattern, i)
    if match:
        print(i)
```

```python
#Email domain filter: 
#Match only emails ending with @gmail.com

emails = ["12.email@example.com", 
          "notvalid-email.com", 
          "user2@domain.org", 
          "praveen@gmail.com",
          "kumar@gmail.org",
          "kumar@gmail.com"]

email_pattern = "@gmail.com+"

for e in emails:
    match = re.search(email_pattern, e)
    if match:
        print(e)
```

```python
Logs = ("ERROR: File not found",
    "INFO: Process started",
    " WARNING: Low memory",
    " ERROR: Disk full")
# Write a regex to match all lines that start with "ERROR"

for l in Logs:
    match = re.search("ER+", l)
    if match:
        print(l)
```
