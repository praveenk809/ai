Stemming => Root word from varients

1. Racing, race, racer => rac
2. driving, drive => driv
3. Gained, gaining, gainful => Gain
4. cars => car
5. caress => care

words ending with S, IES => remove S, IES
words ending with ing => remove ing
wors ending with ed => remove ed

```python
from nltk.stem import PorterStemmer
porter_stemmer = PorterStemmer()
```

```python
from nltk.tokenize import word_tokenize
text = "Very orderly and methodical he looked, with a hand on each knee, and a loud watch ticking a sonorous sermon under his flapped newly bought waist-coat, as though it pitted its gravity and longevity against the levity and evanescence of the brisk fire."
words = word_tokenize(text)
print(words)
```

```python
porter_stemmed = []
for word in words:
    stemmed_word = porter_stemmer.stem(word)
    porter_stemmed.append(stemmed_word)
print(porter_stemmed)
```

## 2. Snowball Stemmer
pip install snowballstemmer

```python
from nltk.stem.snowball import SnowballStemmer
snowball_stemmer = SnowballStemmer("english")

snowball_stemmed = []
for word in words:
    stemmed_word = snowball_stemmer.stem(word)
    snowball_stemmed.append(stemmed_word)
print(snowball_stemmed)
```

```python
# Data Frame 
# word | porter stemmed | snowball stemmed
import pandas as pd
df = pd.DataFrame({"word": words, "porter_stemmed":porter_stemmed, "snowball_stemmed": snowball_stemmed })
df[(df.word!=df.porter_stemmed) | (df.word!=snowball_stemmed)]
```
