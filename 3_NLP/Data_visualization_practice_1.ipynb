```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
s = [3,1,8,0]
l = [6,9,2,4]
s.sort()
l.sort()

print("S numbers:",s)
print("L numbers:",l)

plt.plot(s,l)
plt.grid(True) 

plt.xlabel('S axis')
plt.ylabel('L axis')
plt.title("This is Shreyas and Lauhit")
```

```python
import numpy as np
x = np.linspace( 1, 10, 100)
y = np.sin(x)
z = np.cos(x)

plt.plot(x,y)
plt.plot(x,z)
plt.xlabel('Shreyas')
plt.ylabel('Lauhit')
```

```python

# Linear, Cubic, Log, Exp( Subplats)

# Linear 
x = np.linspace(1,10,100)
y = 4*x
plt.figure(1) # Row =1 
plt.subplot(2,2,1) # grid 2x2, column = 2
plt.title("Linear")
plt.plot(x,y)
plt.subplots_adjust(wspace=.4)
plt.grid()
plt.grid(axis='x',color='red', linestyle='--', linewidth=1, alpha=1)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')


#Cubic

x = np.linspace(1,10,100)
y = x**3
plt.figure(1)
plt.title("Cubic")
plt.subplot(2,2,2)
plt.plot(x,y)
plt.grid()
plt.grid(color='green', linestyle=':', linewidth=1, alpha=0.7)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')


# Exponential
x = np.linspace(1,10,100)
y = np.exp(x)
plt.title('Exp')
plt.figure(2)
plt.subplot(2,2,1)
plt.plot(x,y)
plt.title("Exp")
plt.grid()
plt.grid(color='black', linestyle='-', linewidth=1, alpha=0.7)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')


#Log
x = np.linspace(1,10,100)
y = np.log(x)

plt.figure(2)
plt.subplot(2,2,2)
plt.plot(x,y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(color = 'grey', linestyle ="--", linewidth = .7, alpha = .5)
plt.title("Log")
plt.subplots_adjust(wspace=.5)
```
