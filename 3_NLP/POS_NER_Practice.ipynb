```python
import nltk
nltk.download('maxent_ne_chunker')
nltk.download('maxent_ne_chunker_tab')
```

```python
sentence = "Barack Obama was born in Hawaii and served as President of the United States."
words = nltk.word_tokenize(sentence)
tag_elements = nltk.pos_tag(words)
nltk.ne_chunk(tag_elements)
```

```python
from ipynb.fs.full.NLP_Practice_3 import get_wordnet_pos

wordnet_elements = []

for word, pos in tag_elements:
    wordnet_pos = (word, get_wordnet_pos(pos))
    wordnet_elements.append(wordnet_pos)
wordnet_elements
```

```python
import nltk
nltk.ne_chunk(wordnet_elements)
```
