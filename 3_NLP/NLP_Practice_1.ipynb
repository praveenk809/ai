```python
import requests
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
url = "https://www.gutenberg.org/files/11/11-0.txt"
response = requests.get(url)
words = response.text.split()
len(words),len(response.text)
```

```python
words = [word for word in words if word.lower() not in stop_words]
words
```

```python
from nltk import FreqDist
word_freq = FreqDist(words)
word_freq
```

```python
import seaborn as sns
def plot_word_frequency(words, top_n):
    topn_words = word_freq.most_common(top_n)
    labels = [element[0] for element in topn_words]
    counts = [element[1] for element in topn_words]
    sns.barplot(x=labels, y =counts)
plot_word_frequency(words, 5)
```
